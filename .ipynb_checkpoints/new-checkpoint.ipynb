```python
import numpy as np
import pandas as pd
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
stock = input("stock code:")
data = yf.download(stock, "2008-01-01", "2021-01-18", auto_adjust=True)

data.head()
```
